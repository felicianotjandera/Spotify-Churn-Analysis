```python
import pandas as pd
```

```python
#Load Dataset
file_path = 'spotify_churn_dataset.csv'
df = pd.read_csv(file_path)

#Display Information
df.info()
df.head()
```

```python
#Create a copy of the dataset and remove user_id column
df_clean = df.drop(columns='user_id').copy()

#Convert churn status from 0/1 to No/Yes
df_clean['is_churned'] = df_clean['is_churned'].map({0: 'No', 1: 'Yes'})

#Map country codes to full country names
country_mapping = {'CA': 'Canada', 'DE': 'Germany', 'AU': 'Australia', 'US': 'United States', 'UK': 'United Kingdom', 'IN': 'India', 'FR': 'France', 'PK': 'Pakistan'}
df_clean['country'] = df_clean['country'].map(country_mapping)

#Format column names
columns_rename = {col: col.replace('_', ' ').title() for col in df_clean.columns}
df_clean.rename(columns=columns_rename, inplace=True)

#Display the cleaned dataset
display(df_clean.head())
```

```python
#Export cleaned dataset to CSV
df_clean.to_csv('spotify_churn_clean.csv', index=False)
```
